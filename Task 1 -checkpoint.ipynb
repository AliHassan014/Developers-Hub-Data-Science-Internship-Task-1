# Task 1: Exploring and Visualizing the iris dataset.

## Introduction and Problem statement 
The objective of the task is to explore and visualize the iris dataset . In which we have to load the dataset first and examine it thorough different libraries to collect information regarding the iris dataset. This will be done in the form of graphs such as histogram , scatter plot and box plot.

```python
# Import libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# load the iris dataset
df = pd.read_csv("iris.csv")

# Displaying dataset dimensions 
print("Shape",df.shape)

# Displaying cloumn names 
print("Colums",df.columns)

# first five rows showed
df.head()
```

## Dataset understanding and Description .

## Exploratory data analysis 
This includes the scatter plot , Histogram and Box plot down below .

```python
# Scatter plot showing relationship between sepal length and sepal width
sns.scatterplot(data=df, x="SepalLengthCm", y="SepalWidthCm")
```

```python
# Histogram showing distribution of sepal length 
plt.hist(df["SepalLengthCm"])

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")

plt.show()
```

```python
# Box plot spread and otliers in sepal lentgh by species
sns.boxplot(
    data=df,
    x="Species",
    y="SepalLengthCm"
)

plt.title("Sepal Length by Species")
plt.show()
```
